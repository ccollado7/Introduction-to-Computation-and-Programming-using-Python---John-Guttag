# 3.1 Exhaustive Enumeration

```python
# Find the cube root of a perfect cube

x = int(input("Enter an integer:"))
ans = 0
while(ans**3 < abs(x)):
    ans = ans + 1
if ans**3 != abs(x):
    print(x,"is not a perfect cube")
else:
    if x < 0:
        ans = -ans
    print("Cube root of", x , "is", ans)
```

Cuando se utiliza un loop se debe pensar en utilizar una FUNCION DECREMENTAL apropiada. Esta es una funcion que tiene las siguientes propiedades:

1. Mapea variables del programa en un entero

2. Cuando el loop comienza, su valor es no negativo

2. Cuando el valor es igual o menor a cero el loop debe terminar

4. Su valor debe ir decreciendo cada tiempo a traves del loop

¿Cual es la funcion decremento en el ejemplo anterior?

La funcion decremento es abs(x) - ans**3 ya que mientras esto sea positivo el loop seguira funcionando

A continuacion,considerando el ejemplo anterior, agrego un comentario al inicio del loop para ver cual es el valor de la funcion decremental:

Se observara que se imprima la consigna "El valor de la funcion decremental abs(x) - ans**3 es:", abs(x) - ans**3 en forma repetida con el mismo valor ya que, como se menciono previamente, el loop nunca termina ya que la funcion decremento nunca cambia al no verse modificado el valor de ans

```python
x = int(input("Ingrese un numero entero:"))
ans = 0
while(ans**3 < abs(x)):
    print("El valor de la funcion decremental abs(x) - ans**3 es:", abs(x) - ans**3)
    ans = ans
if ans**3 != abs(x):
    print(x, "No es cubo perfecto")
else:
    if x < 0:
        ans = -ans
    print("La raiz cubica de", x , "es", ans)
```
